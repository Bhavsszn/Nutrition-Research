```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
exe_df = pd.read_csv('exercise_data.csv')
exe_df.shape

food_df = pd.read_csv('food.csv')
food_df
```

```python
unique_user_ids = exe_df['User_ID'].unique()
num_unique_user_ids = len(unique_user_ids)
print(num_unique_user_ids)
```

```python
exe_df.replace({"Gender":{'male':0,'female':1}}, inplace=True)
exe_df.head()
```

```python
exe_df.info()
```

```python
def clean_exercise_type(exercise_type):
    cleaned_type = re.sub("[^a-zA-Z0-9 ]", "", exercise_type)
    return cleaned_type

exe_df["clean_exercise_type"] = exe_df["Exercise_Type"].apply(clean_exercise_type)
exe_df
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 2))
tfidf = vectorizer.fit_transform(exe_df["clean_exercise_type"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def search(exercise_type_name):
# exercise_type_name = "Climbing"
    exercise_type_name = clean_exercise_type(exercise_type_name)
    query_vec = vectorizer.transform([exercise_type_name])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity, -5)[-5:]
    results = exe_df.iloc[indices][::-1]
    return results
```

```python
import ipywidgets as widgets
from IPython.display import display

exercise_input = widgets.Text(
    value = "Basketball",
    description = "Exercises",
    disabled = False

)

exercise_list = widgets.Output()

# def search(title):
#     # Placeholder function for demonstration
#     return widgets.HTML(value=f"<b>Search Results for:</b> {title}")


def on_type(data):
    with exercise_list:
        exercise_list.clear_output()
        title = data["new"]
        if len(title) > 5:
            display(search(title))
        
exercise_input.observe(on_type, names = 'value')

display(exercise_input, exercise_list)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

exe_df['clean_exercise_type'] = exe_df['clean_exercise_type'].fillna('')

# Vectorizing the exercise types
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(exe_df['clean_exercise_type'])

# Calculateing similarity scores between exercise types
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

def recommend_exercise(age, gender, calories_burned, top_n=5):
    filtered_exercises = exe_df[(exe_df['Age'] == age) & (exe_df['Gender'] == gender) &
                                (exe_df['Total_Calories_Burned'] >= calories_burned)]

    # indices of filtered exercises
    indices = filtered_exercises.index.tolist()
    similarity_scores = cosine_sim[indices].sum(axis=0)

    # indices of top recommended exercises
    top_indices = similarity_scores.argsort()[::-1][:top_n]

    #exercise types corresponding to the top indices
    recommended_exercises = exe_df.iloc[top_indices]['Exercise_Type'].tolist()

    return recommended_exercises


age = 45
gender = 1 
calories_burned = 45 
recommended_exercises = recommend_exercise(age, gender, calories_burned)
print("Recommended Exercises:", recommended_exercises)

```

```python
#Food recommendation

# Vectorizing food names
vectorizer_food = TfidfVectorizer(stop_words='english')
tfidf_matrix_food = vectorizer_food.fit_transform(food_df['name'])

# Calculate similarity scores between food items
cosine_sim_food = linear_kernel(tfidf_matrix_food, tfidf_matrix_food)

def recommend_food(age, gender, desired_calories, top_n=5):
    # Filtering food items based on user preferences
    filtered_food = food_df[(food_df['Calories (kcal)'] <= desired_calories)]

    # Indices of filtered food items
    indices_food = filtered_food.index.tolist()
    similarity_scores_food = cosine_sim_food[indices_food].sum(axis=0)

    top_indices_food = similarity_scores_food.argsort()[::-1][:top_n]

    # Food items corresponding to the top indices
    recommended_food = food_df.iloc[top_indices_food]['name'].tolist()

    return recommended_food

age = 45
gender = 1 
desired_calories = 900  # Adjust this according to user preference
recommended_food = recommend_food(age, gender, desired_calories)
print("Recommended Food:", recommended_food)
```
